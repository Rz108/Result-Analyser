```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FakeScore.csv')

df = df.rename(columns={'Unnamed: 0': 'Name'})

display(df)
```

```python
#function
def categorize_score(score):

    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "E"
```

```python
math = df['Math'].tolist()
result_for_math = []
for i in math:
    grade = categorize_score(i)
    #print(grade)
    result_for_math += grade
print(result_for_math)

english = df['English'].tolist()
result_for_english = []
for i in english:
    grade = categorize_score(i)
    #print(grade)
    result_for_english += grade
print(result_for_english)


sci = df['Science'].tolist()
result_for_sci = []
for i in sci:
    grade = categorize_score(i)
    #print(grade)
    result_for_sci += grade
print(result_for_sci)
```

```python
df['Math_cat'] = result_for_math

df['English_cat'] = result_for_english

df['Sci_cat'] = result_for_sci
df
```

```python
df.drop(['Name','Math','English','Science'], axis=1)
```

```python
df.to_csv('FakeScoreCat.csv')
```
