```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('subScore.csv')

#df = df.rename(columns={'Unnamed: 0': 'Name'})

display(df)
```

```python
#function
def categorize_score(score):
    if score == 5:
        return "high"
    elif score == 4:
        return "middle"
    elif score == 3:
        return "middle"
    else:
        return "low"
    
```

```python
math = df['Numbers and their operations'].tolist()
result_for_math = []
for i in math:
    grade = categorize_score(i)
    result_for_math.append(grade)
print(result_for_math)


math = df['Ratio and proportion'].tolist()
result_for_math2 = []
for i in math:
    grade = categorize_score(i)
    result_for_math2.append(grade)
print(result_for_math2)


math = df['Percentage'].tolist()
result_for_math3 = []
for i in math:
    grade = categorize_score(i)
    result_for_math3.append(grade)
print(result_for_math3)


math = df['Rate and speed'].tolist()
result_for_math2 = []
for i in math:
    grade = categorize_score(i)
    result_for_math2.append(grade)
print(result_for_math2)


```

```python
for i in df.columns:
    math = df[i].tolist()
    result_for_math = []
    for j in math:
        grade = categorize_score(j)
        result_for_math.append(grade)
df[i] = result_for_math

```
